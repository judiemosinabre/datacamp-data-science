```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like

# Filter and store 1990s movies in nineties_movies
nineties_subset = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)]

nineties_movies = nineties_subset[nineties_subset['type'] == 'Movie']
nineties_movies
```

```python
# Filter 1990s movie duration
nm_duration = nineties_movies['duration']

# Use Histogram to display frequent movie duration
plt.hist(nm_duration)
plt.title('Movie Duration in 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()

# Frequent Duration based from Histogram
duration = 100
print(duration)
```

```python
# Filter 1990s action movies 
nm_action_movies = nineties_movies[nineties_movies['genre'] == 'Action'] 


short_movie_count = 0

# Create a for loop to iterate in each row to look for row['duration'] less than 90
for label, row in nm_action_movies.iterrows():
    if row['duration'] < 90:
        short_movie_count += 1

# Display Short Movies
print(short_movie_count)
```
